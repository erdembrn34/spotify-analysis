```python
# 1. Kütüphaneleri yükle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Notebook içinde grafikler için
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('SpotifyFeatures.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df = df.dropna(subset=['track_name'])
```

```python
plt.figure(figsize=(12,6))
top_genres = df['genre'].value_counts().head(20)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('En Popüler 20 Tür')
plt.xlabel('Şarkı Sayısı')
plt.ylabel('Tür')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['popularity'], bins=30, kde=True, color='skyblue')
plt.title('Popülerlik Dağılımı')
plt.xlabel('Popülerlik')
plt.ylabel('Sıklık')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='tempo', y='energy', hue='popularity', data=df.sample(10000), palette='coolwarm', alpha=0.6)
plt.title('Tempo ve Energy vs Popülerlik (Örnek 10k)')
plt.xlabel('Tempo')
plt.ylabel('Energy')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='danceability', y='popularity', data=df.sample(10000), alpha=0.6, color='green')
plt.title('Danceability vs Popülerlik (Örnek 10k)')
plt.xlabel('Danceability')
plt.ylabel('Popülerlik')
plt.show()
```

```python
top_artists = df.groupby('artist_name')['popularity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='magma')
plt.title('En Popüler 10 Sanatçı (Ortalama Popülerlik)')
plt.xlabel('Ortalama Popülerlik')
plt.ylabel('Sanatçı')
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr_features = ['popularity','danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
sns.heatmap(df[corr_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Sayısal Özellikler Arası Korelasyon')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='loudness', y='popularity', data=df.sample(10000), alpha=0.5)
plt.title('Loudness vs Popülerlik (Örnek 10k)')
plt.xlabel('Loudness (dB)')
plt.ylabel('Popülerlik')
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_10_genres = df['genre'].value_counts().head(10).index
sns.boxplot(x='genre', y='popularity', data=df[df['genre'].isin(top_10_genres)])
plt.title('Türlere Göre Popülerlik Dağılımı (Top 10 Tür)')
plt.xlabel('Tür')
plt.ylabel('Popülerlik')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='danceability', y='energy', data=df.sample(10000), alpha=0.5)
plt.title('Danceability vs Energy (Örnek 10k)')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_genres = df['genre'].value_counts().head(20)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('En Popüler 20 Tür')
plt.xlabel('Şarkı Sayısı')
plt.ylabel('Tür')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='tempo', y='energy', hue='popularity', data=df.sample(10000), palette='coolwarm', alpha=0.6)
plt.title('Tempo ve Energy vs Popülerlik (Örnek 10k)')
plt.xlabel('Tempo')
plt.ylabel('Energy')
plt.show()
```
